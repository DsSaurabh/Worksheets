### 12. Considering the 'Price' column as target variable which is to be predicted, how will you treat the missing values? Give a brief explanation and treat the missing values in 'Price' column.

```python
df['Price'].isnull().sum()
```

```python
(df['Price'].isnull().sum()/len(df)) * 100
```

Explanation:
    Price column is a target column and in this column more than 20% null value are present.So we are not dropping all the null
    value of Price columns because if we drop all the null value of the price columns more data is lost so drop more data 
    it's not good way.

```python
#price column missing value fillup by mean 
df['Price'].fillna(df['Price'].mean(),inplace=True)
```
